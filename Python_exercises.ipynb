```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
population_inferred = pd.read_csv('Canada_population_2024.csv',
                        sep=',',
                        skiprows=1,
                        header=None,
                        names=['province', 'date', 'q1_2023', 'q2_2023', 'q3_2023', 'q4_2023', 'q1_2024'],
                        parse_dates=['date'],
                        dtype={
                            'province': 'string',
                            'q1_2023' : 'int64',
                            'q2_2023' : 'int64',
                            'q3_2023' : 'int64',
                            'q4_2023' : 'int64',
                            'q1_2024' : 'int64',
                            },
                        thousands=','
                        )
population_inferred
```

```python
population_inferred.dtypes
```

```python
population_inferred['2023_2024_change'] = population_inferred['q1_2024'] - population_inferred['q1_2023']
population_inferred
```

```python
population_inferred.set_index('province', inplace = True)
population_inferred
```

```python
population_inferred.at['Canada', '2023_2024_change']
```

```python
population_inferred.drop('date', axis = 1, inplace = True)
population_inferred
```

```python
population_inferred = population_inferred.drop(['Canada'])
population_inferred
```

```python
def percent_change(quarters):
    q1, q2 = quarters
    change = (q2 - q1)/q1*100
    return change
```

```python
population_inferred['2023_2024_change_percent'] = population_inferred[['q1_2023', 'q1_2024']].apply(percent_change, axis = 1)
population_inferred
```

```python
population_inferred['2023_2024_change_percent'] = population_inferred['2023_2024_change_percent'].map(lambda x: '%.2f' % x)
population_inferred
```

```python
population_inferred.sort_values('province')
```

```python
population_inferred.sort_values('2023_2024_change_percent', ascending = False)
```

```python
population_inferred.dtypes
```

```python
population_inferred['2023_2024_change_percent'] = population_inferred['2023_2024_change_percent'].astype(float)
population_inferred.dtypes
```

```python
population_inferred['top_change'] = \
        np.where(population_inferred['2023_2024_change_percent'] > 3, 'above_3%_change', 'below_3%_change')
population_inferred
```

```python
groupby_top_change = \
population_inferred.groupby('top_change').agg({'2023_2024_change':['min','max'], '2023_2024_change_percent':['min','max']})
groupby_top_change
```

```python
groupby_top_change.columns = groupby_top_change.columns.to_flat_index()
groupby_top_change
```

```python
groupby_top_change.columns = ["_".join(a) for a in groupby_top_change.columns.to_flat_index()]
groupby_top_change
```

```python
df1 = pd.DataFrame({'left_key': ['one', 'two', 'three', 'four'],
                    'value': [1,2,3,4]})
df2 = pd.DataFrame({'right_key': ['one', 'two', 'three', 'four'],
                    'value': [11,21,31,41]})
```

```python
df1.merge(df2,
          left_on='left_key',
          right_on='right_key',
          suffixes=('_left', '_right'),
          validate='one_to_one')
```
